```python
import pandas as pd

df = pd.DataFrame({
    'date': ['2024-08-01', '2024-08-02', '2024-08-03', '2024-08-01', '2024-08-02'],
    'region': ['North', 'North', 'North', 'South', 'South'],
    'sales_amount': [100, 150, 200, 50, 75]
})

df.head()
```

```python
df['row_num'] = df.groupby('region').cumcount()+1
```

```python
df['rank'] = df.groupby('region')['sales_amount'].rank(method='min' , ascending=True)

df['dense_rank'] = df.groupby('region')['sales_amount'].rank(method='dense' , ascending=True)
df
```

```python
df['quantile'] =pd.qcut(df['sales_amount'], 3)
```

```python
df['cumulative_sum'] = df.groupby('region')['sales_amount'].cumsum()
```

```python
df['previous_day_sales'] = df.groupby('region')['sales_amount'].shift()
```

```python
df['next_day_sales'] = df.groupby('region')['sales_amount'].shift(-1)
```

```python
df
```

```python
df['fisrt_value'] = df.groupby('region')['sales_amount'].transform('first')
```

```python
df['last_value'] = df.groupby('region')['sales_amount'].transform('last')
```

```python
df['rolling_mean'] = df.groupby('region')['sales_amount'].rolling(window=2).mean().reset_index(level=0, drop = True)
```
