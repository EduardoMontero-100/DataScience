```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
np.random.seed(123)
v1 = pd.Series(np.random.normal(1, 10, 1000), name= 'v1')
v2 = pd.Series(2*v1 +np.random.normal(60, 15, 1000), name = 'v2')
```

```python
plt.figure()
plt.hist(v1, alpha= 0.7, bins = np.arange(-50, 150, 5), label = 'v1'); # bins ==> q bins
plt.hist(v2, alpha= 0.7, bins = np.arange(-50, 150, 5), label = 'v2'); 
plt.legend();
```

```python
plt.figure()
plt.hist([v1, v2], histtype='barstacked')
v3 = np.concatenate([v1,v2])
sns.kdeplot(v3);
```

```python
plt.style.available
```

```python
# escojemos este color palette
plt.style.use('seaborn-dark')
```

```python
plt.figure();
sns.distplot(v3, hist_kws = {'color': 'Teal'}, kde_kws = {'color':'Navy'});
```

```python
sns.jointplot(v1, v2, alpha = 0.6);
```

```python
grid = sns.jointplot(v1, v2, alpha= 0.7)
grid.ax_joint.set_aspect('equal')
```

```python
sns.jointplot(v1, v2, kind= 'hex')
```

```python
sns.set_style('darkgrid')
# kernel density estimations
sns.jointplot(v1, v2, kind= 'kde', space = 0)
```
