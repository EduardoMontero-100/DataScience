## Lineplots

```python
import numpy as np
import matplotlib.pyplot as plt

linear_data = np.array([1,2,3,4,5,6,7,8])
quadratic_data = linear_data**2

plt.figure()
plt.plot(linear_data, '-o', quadratic_data, '-o')
plt.plot([22,44,55], '--r')
plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.legend(['Baseline', 'Competition', 'Us'])
# adding transparency
plt.gca().fill_between(range(len(linear_data)),
                       linear_data, quadratic_data,
                       facecolor = 'blue', alpha=0.25)
```

```python
plt.figure(figsize=(9,5))

observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')

plt.plot(observation_dates, linear_data, '-o',
         observation_dates, quadratic_data, '-o')
```

Chequear como trabaja la funciÃ³n map [map](https://www.geeksforgeeks.org/python-map-function/)

```python
import pandas as pd

# The returned value from map() (map object) then can be passed to functions like list() 
# (to create a list), set() (to create a set) .
plt.figure(figsize=(10,5))
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates))
plt.plot(observation_dates, linear_data, '-o',
         observation_dates, quadratic_data, '-o')

#GCA (get current axis)
x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)

plt.subplots_adjust(bottom=0.25)

ax = plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Units')
ax.set_title('Quadratic vs. Linear performance')
# podemos conectar plt con latex
ax.set_title('Quadratic ($x^2$) vs. Linear ($x$) performance')
```

## Bar Charts

```python
import matplotlib.pyplot as plt
import numpy as np
from random import randint

linear_data = np.array([1,2,3,4,5,6,7,8])
quadratic_data = linear_data**2
xvals = range(len(linear_data))
new_xvals = []

linear_err = [randint(0,15) for x in range(len(linear_data))]

for item in xvals:
    new_xvals.append(item + .3)

plt.figure()
plt.bar(xvals, linear_data, width=0.3)
plt.bar(new_xvals, quadratic_data, width=0.3, color = 'red')
plt.bar(xvals, linear_data, width=0.3, yerr = linear_err)
```

```python
# stack bars
plt.figure()
xvals = range(len(linear_data))
plt.bar(xvals, linear_data, width=0.3, color ='r')
plt.bar(xvals, quadratic_data, width=0.3, color ='b',bottom=linear_data)
```

```python
# stack bars
plt.figure()
xvals = range(len(linear_data))
plt.barh(xvals, linear_data, height=0.3, color ='r')
plt.barh(xvals, quadratic_data, height=0.3, color ='b',left=linear_data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

language = ['Python', 'SQL', 'Java', 'C++', 'Javascript']
pos = np.arange(len(language))
popularity = [56, 39, 34, 34, 29]

bars = plt.bar(pos, popularity, align='center', linewidth=0,color = 'lightslategrey')
# make one bar, the python bar, a contrasting color
bars[0].set_color('#1F77B4')

# soften all labels by turning grey
plt.xticks(pos, language, alpha=0.8)
#plt.ylabel('% Popularity', alpha=0.8)
plt.title('Top 5 Languages for Math and \nData Popularity on Stack Overflow', alpha = 0.9)
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft= 'off', labelbottom  ='on')
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# direct label each bar with Y axis values
for bar in bars:
    height = bar.get_height()
    plt.gca().text(bar.get_x() + bar.get_width()/2, bar.get_height() - 5, str(int(height)) + '%', 
                 ha='center', color='w', fontsize=11)
```

## Histograms

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, sharex = True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0, len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc = 0.0, scale =1.0, size= sample_size)
    axs[n].hist(sample)
    axs[n].set_title(f'{sample_size}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, sharex = True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0, len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc = 0.0, scale =1.0, size= sample_size)
    axs[n].hist(sample, bins = 100)
    axs[n].set_title(f'{sample_size}')
```

## Box and Whistle Plot

```python
import pandas as pd
import numpy as np
normal_sample = np.random.normal(loc = 0.0, scale = 1.0, size = 10000) 
random_sample = np.random.random(size = 10000)
gamma_sample1 = np.random.gamma(1, size = 10000)
gamma_sample2 = np.random.gamma(2, size = 10000)

df = pd.DataFrame({'normal': normal_sample,
                   'random': random_sample,
                   'gamma1': gamma_sample1,
                   'gamma2': gamma_sample2
                   })
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
plt.figure()
_ = plt.boxplot(df['normal'], whis =  2.5)
```

```python
plt.clf()
_ = plt.boxplot([df['normal'], df['random'],df['gamma2']], whis=2.5)
```

```python
plt.figure()
_ = plt.hist(df['gamma2'], bins = 100)
```

```python
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

plt.figure()
plt.boxplot([df['normal'], df['random'], df['gamma2']], whis = 3)
ax2 = mpl_il.inset_axes(plt.gca(), width = '60%', height= '40%', loc = 2) 
ax2.hist(df['gamma2'], bins = 100)
ax2.margins(x=0.8)  
ax2.yaxis.tick_right()
```
